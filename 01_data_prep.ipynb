# Load shapefile

```python
import geopandas as gpd

# Set pyogrio as default enigne
import pyogrio
gpd.options.io_engine = "pyogrio"
```

```python
file_path = 'HomeHill_GNSS_2023_shp/HomeHill2023_combined_GNSS_MGA2020.shp'
gdf = gpd.read_file(file_path)
gdf.head()
```

```python
gdf.crs
```

The input data is in a compound coordinate reference system (CRS):

* `MGA2020 Zone 55` (EPSG:7855) for the horizontal
* `AHD (Tasmania) Height` for the vertical

For simplicity I'll reproject into just MGA2020 so it's a single CRS. Note that geopandas still keeps the geometry as a POINT Z (i.e. a point with some altitude). So we are losing the information on exactly what type of altitude here we mean. We'll also drop all the other columns and just take `geometry` and `OrthoHeigh` which we'll rename to `elevation`.

```python
gdf_mga2020 = gdf.to_crs(epsg="7855")
gdf_mga2020.head()
```

```python
crs = gdf_mga2020.crs
crs
```

```python
survey_points_MGA2020 = gpd.GeoDataFrame(gdf_mga2020['geometry'], crs = crs)
survey_points_MGA2020['elevation'] = gdf_mga2020['OrthoHeigh']
survey_points_MGA2020
```

We'll save the output as another shapefile and in geoparquet format if we can.

```python
# Save to shapefile

survey_points_MGA2020.to_file('py_output/survey_points_MGA2020.shp')
```

```python
# Save to geoparquet

survey_points_MGA2020.to_parquet('py_output/survey_points_MGA2020.parquet')
```
